## 1. Business Problem & Objective

### Background

New York City's restaurant landscape is highly competitive and growing rapidly. A large segment of the city's population — students, young professionals, and busy households — relies on food delivery services for daily meals. FoodHub operates as a food aggregator platform, connecting customers to multiple restaurants through a single smartphone application.

The FoodHub model works as follows:
- A customer places an order through the app.
- The restaurant confirms and prepares the order.
- A FoodHub delivery agent picks up the food and delivers it to the customer.
- The customer rates the experience in the app.
- FoodHub earns a commission margin from each completed restaurant order.

### Business Objective

The leadership team at FoodHub seeks to leverage historical order data to:

1. **Understand demand patterns** — which restaurants, cuisines, and time periods drive the most orders.
2. **Identify operational bottlenecks** — where are delays occurring in preparation or delivery?
3. **Quantify platform revenue** — assess the commission structure and top revenue-generating segments.
4. **Improve customer satisfaction** — identify what factors correlate with high ratings and what leads to poor feedback.
5. **Drive strategic decisions** — support marketing, partnerships, and operational improvement.

### Data Dictionary

| Column | Description |
|---|---|
| `order_id` | Unique identifier for each order |
| `customer_id` | Unique identifier for each customer |
| `restaurant_name` | Name of the restaurant that fulfilled the order |
| `cuisine_type` | Type of cuisine ordered |
| `cost_of_the_order` | Total cost of the order in USD |
| `day_of_the_week` | Whether order was placed on a Weekday or Weekend |
| `rating` | Customer rating (1–5 stars), or 'Not given' |
| `food_preparation_time` | Minutes from order confirmation to delivery agent pickup |
| `delivery_time` | Minutes from pickup to customer drop-off |

```python
# ── Standard library ──────────────────────────────────────────────────────────
import warnings
warnings.filterwarnings('ignore')

# ── Data manipulation ─────────────────────────────────────────────────────────
import numpy as np
import pandas as pd

# ── Visualisation ─────────────────────────────────────────────────────────────
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
from scipy import stats

# ── Global plot aesthetics ────────────────────────────────────────────────────
PALETTE_MAIN   = 'Blues_r'
ACCENT_COLOR   = '#2563EB'   # brand blue
HIGHLIGHT      = '#F59E0B'   # amber for callouts
NEUTRAL        = '#64748B'   # slate grey
FONT_TITLE     = {'fontsize': 14, 'fontweight': 'bold', 'pad': 12}
FONT_AXIS      = {'fontsize': 11}

sns.set_theme(style='whitegrid', palette='muted', font_scale=1.05)
plt.rcParams.update({
    'figure.dpi'       : 120,
    'axes.spines.top'  : False,
    'axes.spines.right': False,
    'figure.facecolor' : 'white',
})

pd.set_option('display.float_format', '{:.2f}'.format)
pd.set_option('display.max_columns', 20)

print('Environment configured successfully.')
```

## 4. Data Cleaning & Missing Value Treatment

```python
# Work on a copy to preserve the original data
df = df_raw.copy()

# ── Missing values ────────────────────────────────────────────────────────────
missing = df.isnull().sum()
print('Missing values per column:')
print(missing[missing > 0] if missing.any() else 'No missing values detected.')

# ── Duplicate rows ────────────────────────────────────────────────────────────
duplicates = df.duplicated().sum()
print(f'\nDuplicate rows: {duplicates}')
```

```python
# ── Clean the 'rating' column ─────────────────────────────────────────────────
# 'Not given' is a valid business state (customer chose not to rate).
# We keep it as NaN for numeric operations, but add a boolean flag for segmentation.

df['rating_given']  = df['rating'].ne('Not given')   # True if rated
df['rating_numeric'] = pd.to_numeric(
    df['rating'].replace('Not given', np.nan), errors='coerce'
)

# ── Derived features ──────────────────────────────────────────────────────────
df['total_fulfillment_time'] = df['food_preparation_time'] + df['delivery_time']
df['exceeds_60_min']         = df['total_fulfillment_time'] > 60

# Commission revenue per order (per company's pricing rules)
df['commission_revenue'] = df['cost_of_the_order'].apply(
    lambda cost: cost * 0.25 if cost > 20
    else (cost * 0.15 if cost > 5 else 0)
)

print('Data cleaning complete.')
print(f'Final dataset: {df.shape[0]:,} rows × {df.shape[1]} columns')
```

### Data Quality Summary

The dataset is in good structural shape with no duplicate rows and no truly missing values in core transaction fields. Key observations:

- **Rating** is the only column with incomplete data — a large share of orders carry `'Not given'` as the rating value. This is a business problem (low engagement), not a data error. We treat these as `NaN` for numerical analysis and preserve them as a separate category for segmentation.
- **All time and cost fields** are fully populated with no anomalies detected.
- **Three new derived columns** have been added to support downstream analysis: `total_fulfillment_time`, `exceeds_60_min`, and `commission_revenue`.

### Q2: Distribution of All Variables

```python
# ── Helper: annotate bars with counts ─────────────────────────────────────────
def annotate_bars(ax, fmt='{:.0f}', fontsize=9, color='#374151'):
    """Add value labels to the end of each bar in a bar/count plot."""
    for patch in ax.patches:
        width  = patch.get_width()
        height = patch.get_height()
        # Horizontal bar
        if width > 0 and height < width:
            ax.text(
                width + ax.get_xlim()[1] * 0.005,
                patch.get_y() + height / 2,
                fmt.format(width), va='center', ha='left',
                fontsize=fontsize, color=color
            )
        # Vertical bar
        elif height > 0:
            ax.text(
                patch.get_x() + patch.get_width() / 2,
                height + ax.get_ylim()[1] * 0.005,
                fmt.format(height), va='bottom', ha='center',
                fontsize=fontsize, color=color
            )
```

```python
# ── 5.1 Cuisine Type Distribution ─────────────────────────────────────────────
cuisine_counts = df['cuisine_type'].value_counts()

fig, ax = plt.subplots(figsize=(9, 6))
bars = ax.barh(
    cuisine_counts.index, cuisine_counts.values,
    color=[ACCENT_COLOR if i < 3 else NEUTRAL for i in range(len(cuisine_counts))]
)
ax.set_title('Cuisine Type — Order Volume', **FONT_TITLE)
ax.set_xlabel('Number of Orders', **FONT_AXIS)
ax.set_ylabel('')
annotate_bars(ax)
ax.invert_yaxis()
plt.tight_layout()
plt.show()
```

**What this chart shows:** The number of orders placed for each cuisine type, ranked from most to least popular.

**Why it matters:** American, Japanese, and Italian cuisines are dominating the platform, together accounting for the large majority of all orders. Many other cuisines are barely used.

**What FoodHub should do:** Double down on marketing and restaurant recruitment in the top-3 cuisines. For niche cuisines, run targeted promotions to test whether low demand is due to limited supply of good restaurants or simply low customer interest.

**What this chart shows:** The split between orders placed on weekdays vs. weekends.

**Why it matters:** Weekends generate significantly more orders. This creates predictable demand spikes that must be operationally planned for.

**What FoodHub should do:** Ensure adequate delivery agent coverage on weekends to prevent delays. Consider weekday promotions to balance demand more evenly, reducing pressure on weekend operations.

## 6. Bivariate & Multivariate Analysis

### Q8: Exploring Relationships Between Key Variables

```python
# ── 6.1 Delivery Time vs Customer Rating ──────────────────────────────────────
# Only use rows where a rating was actually given
rated_df = df[df['rating_given']].copy()
rated_df['rating'] = rated_df['rating'].astype(str)

fig, ax = plt.subplots(figsize=(9, 5))
sns.boxplot(
    data=rated_df, x='rating', y='delivery_time',
    palette=['#DBEAFE', '#93C5FD', ACCENT_COLOR], ax=ax, order=['3', '4', '5']
)
ax.set_title('Delivery Time by Customer Rating', **FONT_TITLE)
ax.set_xlabel('Star Rating', **FONT_AXIS)
ax.set_ylabel('Delivery Time (minutes)', **FONT_AXIS)
plt.tight_layout()
plt.show()
```

**What this chart shows:** Whether customers who waited longer for delivery gave lower ratings.

**Why it matters:** If long deliveries lead to poor ratings, fixing delivery speed is directly linked to customer satisfaction scores.

**What FoodHub should do:** Investigate whether the median delivery times differ significantly across rating groups. If 3-star orders have noticeably longer delivery times, the company should set a delivery time SLA and use it to guide staffing and routing decisions.

```python
# ── 6.2 Preparation Time by Cuisine Type ─────────────────────────────────────
fig, ax = plt.subplots(figsize=(11, 6))
cuisine_prep_order = (
    df.groupby('cuisine_type')['food_preparation_time']
    .median()
    .sort_values(ascending=False)
    .index
)
sns.boxplot(
    data=df, x='cuisine_type', y='food_preparation_time',
    order=cuisine_prep_order, color=ACCENT_COLOR, ax=ax
)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha='right')
ax.set_title('Food Preparation Time by Cuisine Type', **FONT_TITLE)
ax.set_xlabel('Cuisine Type', **FONT_AXIS)
ax.set_ylabel('Preparation Time (minutes)', **FONT_AXIS)
plt.tight_layout()
plt.show()
```

**What this chart shows:** Which cuisines take the longest to prepare, and how consistent preparation times are within each cuisine type.

**Why it matters:** Cuisines with high or highly variable preparation times create delivery scheduling challenges and may frustrate customers waiting longer than expected.

**What FoodHub should do:** Set cuisine-specific preparation time benchmarks. Communicate estimated wait times clearly in the app for cuisines with longer preparation, so customers are not disappointed.

```python
# ── 6.3 Delivery Time — Weekday vs Weekend ─────────────────────────────────────
fig, ax = plt.subplots(figsize=(8, 5))
sns.boxplot(
    data=df, x='day_of_the_week', y='delivery_time',
    palette=[ACCENT_COLOR, NEUTRAL], ax=ax
)
ax.set_title('Delivery Time — Weekday vs Weekend', **FONT_TITLE)
ax.set_xlabel('Day Type', **FONT_AXIS)
ax.set_ylabel('Delivery Time (minutes)', **FONT_AXIS)
plt.tight_layout()
plt.show()

# Statistical comparison
weekday_del = df[df['day_of_the_week'] == 'Weekday']['delivery_time']
weekend_del = df[df['day_of_the_week'] == 'Weekend']['delivery_time']
t_stat, p_val = stats.ttest_ind(weekday_del, weekend_del)

print(f'Weekday mean delivery : {weekday_del.mean():.2f} min')
print(f'Weekend mean delivery : {weekend_del.mean():.2f} min')
print(f'Difference            : {weekend_del.mean() - weekday_del.mean():.2f} min')
print(f'T-test p-value        : {p_val:.4f} ({"statistically significant" if p_val < 0.05 else "not significant"})')
```

**What this chart shows:** Whether deliveries take longer on weekends compared to weekdays.

**Why it matters:** If weekend delivery times are higher, this would confirm that higher demand is straining the delivery network — a direct operational risk to customer satisfaction.

**What FoodHub should do:** If the difference is statistically significant, invest in surge staffing on weekends. If the difference is small and the test is not significant, the current operation scales well with demand.

```python
# ── 6.4 Order Cost vs Customer Rating ─────────────────────────────────────────
fig, ax = plt.subplots(figsize=(9, 5))
rated_df = df[df['rating_given']].copy()
rated_df['rating'] = rated_df['rating'].astype(str)

sns.boxplot(
    data=rated_df, x='rating', y='cost_of_the_order',
    palette=['#DBEAFE', '#93C5FD', ACCENT_COLOR], ax=ax, order=['3', '4', '5']
)
ax.set_title('Order Cost by Customer Rating', **FONT_TITLE)
ax.set_xlabel('Star Rating', **FONT_AXIS)
ax.set_ylabel('Order Cost (USD)', **FONT_AXIS)
plt.tight_layout()
plt.show()
```

**What this chart shows:** Whether customers who spend more tend to rate the experience higher or lower.

**Why it matters:** If high-spend customers rate lower, there may be unmet expectations around premium orders (e.g., presentation, temperature, accuracy). If they rate higher, premium positioning should be further developed.

**What FoodHub should do:** Use this relationship to decide whether quality assurance measures should be applied specifically to high-value orders.

```python
# ── 6.5 Average Delivery Time by Cuisine Type ──────────────────────────────────
cuisine_del = (
    df.groupby('cuisine_type')['delivery_time']
    .mean()
    .sort_values(ascending=True)
)

fig, ax = plt.subplots(figsize=(9, 6))
colors = [ACCENT_COLOR if v == cuisine_del.max() else
          HIGHLIGHT if v == cuisine_del.min() else NEUTRAL
          for v in cuisine_del.values]
ax.barh(cuisine_del.index, cuisine_del.values, color=colors)
ax.set_title('Average Delivery Time by Cuisine Type', **FONT_TITLE)
ax.set_xlabel('Average Delivery Time (minutes)', **FONT_AXIS)
ax.set_ylabel('')
annotate_bars(ax, fmt='{:.1f}')
plt.tight_layout()
plt.show()
```

**What this chart shows:** The average time it takes for orders to be delivered, broken down by cuisine type.

**Why it matters:** If certain cuisines consistently have longer delivery times, it may point to restaurant clustering issues, geographic concentration, or specific kitchen preparation bottlenecks affecting logistics.

**What FoodHub should do:** Investigate the outlier cuisines at both ends of the spectrum. For slow-delivery cuisines, explore whether restaurant locations, agent routing, or preparation hand-off timing can be optimised.

```python
# ── 6.6 Correlation Heatmap ────────────────────────────────────────────────────
corr_cols = ['cost_of_the_order', 'food_preparation_time', 'delivery_time',
             'total_fulfillment_time', 'commission_revenue']

corr_matrix = df[corr_cols].corr()

fig, ax = plt.subplots(figsize=(8, 6))
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
sns.heatmap(
    corr_matrix, annot=True, fmt='.2f', cmap='Blues',
    mask=mask, ax=ax, linewidths=0.5, linecolor='white',
    vmin=-1, vmax=1
)
ax.set_title('Correlation Matrix — Numeric Variables', **FONT_TITLE)
plt.tight_layout()
plt.show()
```

**What this chart shows:** The strength of the statistical relationship between each pair of numeric variables.

**Why it matters:** Strong correlations reveal which variables move together, helping prioritise which factors to investigate as root causes vs. symptoms.

**What FoodHub should do:** Pay attention to the relationship between `cost_of_the_order` and `commission_revenue` (by design, highly correlated). More importantly, assess whether `food_preparation_time` and `delivery_time` are independent — if they are, delays in one cannot be blamed on the other.

```python
# ── 6.7 Average Rating of Top 5 Restaurants ───────────────────────────────────
top5_names = df['restaurant_name'].value_counts().head(5).index

top5_avg_rating = (
    df[df['restaurant_name'].isin(top5_names)]
    .groupby('restaurant_name')['rating_numeric']
    .agg(avg_rating='mean', rating_count='count')
    .sort_values('avg_rating')
)

fig, ax = plt.subplots(figsize=(9, 5))
bars = ax.barh(top5_avg_rating.index, top5_avg_rating['avg_rating'],
               color=ACCENT_COLOR)
ax.set_xlim(0, 5.5)
ax.axvline(4.0, color=HIGHLIGHT, linestyle='--', label='Rating threshold (4.0)')
ax.set_title('Average Rating — Top 5 Busiest Restaurants', **FONT_TITLE)
ax.set_xlabel('Average Rating (rated orders only)', **FONT_AXIS)
ax.legend()
annotate_bars(ax, fmt='{:.2f}')
plt.tight_layout()
plt.show()

print(top5_avg_rating)
```

**What this chart shows:** How highly rated the five busiest restaurants are — comparing popularity with quality.

**Why it matters:** A popular restaurant with a low rating is a risk. It means many customers are choosing it but not enjoying the experience — which could drive churn.

**What FoodHub should do:** Conduct a deeper investigation into any top restaurant rated below 4.0. High order volume with low satisfaction is a churn risk and potentially a reputational risk for the platform.

### Q9: Restaurants Eligible for Promotional Offer

```python
# Criteria: rating count > 50 AND average rating > 4.0
restaurant_ratings = (
    df.groupby('restaurant_name')['rating_numeric']
    .agg(rating_count='count', avg_rating='mean')
    .reset_index()
)

eligible = (
    restaurant_ratings[
        (restaurant_ratings['rating_count'] > 50) &
        (restaurant_ratings['avg_rating'] > 4.0)
    ]
    .sort_values('avg_rating', ascending=False)
    .reset_index(drop=True)
)

eligible.index += 1  # 1-based ranking
eligible.columns = ['Restaurant', 'Ratings Received', 'Average Rating']

print(f'Restaurants eligible for promotional offer: {len(eligible)}\n')
eligible
```

**Answer:** The restaurants listed above meet both criteria: they have received enough ratings to be statistically meaningful (>50 rated orders) **and** their average rating exceeds 4.0. These restaurants represent FoodHub's highest-quality, highest-engagement partners — the ideal candidates for featured placement in advertising campaigns.

## 8. Advanced Analysis — Revenue, Demand & Bottlenecks

```python
# ── 8.1 Revenue by Cuisine Type ────────────────────────────────────────────────
cuisine_revenue = (
    df.groupby('cuisine_type')['commission_revenue']
    .sum()
    .sort_values(ascending=True)
)

fig, ax = plt.subplots(figsize=(9, 6))
colors = [ACCENT_COLOR if i >= len(cuisine_revenue) - 3 else NEUTRAL
          for i in range(len(cuisine_revenue))]
ax.barh(cuisine_revenue.index, cuisine_revenue.values, color=colors)
ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'${x:,.0f}'))
ax.set_title('Total Commission Revenue by Cuisine Type', **FONT_TITLE)
ax.set_xlabel('Total Revenue (USD)', **FONT_AXIS)
annotate_bars(ax, fmt='${:,.0f}')
plt.tight_layout()
plt.show()
```

**What this chart shows:** The total commission revenue FoodHub earns from each cuisine category.

**What FoodHub should do:** Prioritise partnerships and marketing investment in the top-3 revenue-generating cuisines. These are where the business is most dependent for its income.

```python
# ── 8.2 Weekday vs Weekend — Cuisine Demand Heatmap ───────────────────────────
demand_pivot = (
    df.groupby(['cuisine_type', 'day_of_the_week'])
    .size()
    .unstack(fill_value=0)
)
# Normalise by row to show % split within each cuisine
demand_pivot_pct = demand_pivot.div(demand_pivot.sum(axis=1), axis=0) * 100
demand_pivot_pct = demand_pivot_pct.sort_values('Weekend', ascending=False)

fig, ax = plt.subplots(figsize=(8, 7))
sns.heatmap(
    demand_pivot_pct, annot=True, fmt='.1f', cmap='Blues',
    linewidths=0.5, linecolor='white', ax=ax, vmin=0, vmax=100
)
ax.set_title('Weekend vs Weekday Order Share (%) by Cuisine', **FONT_TITLE)
ax.set_xlabel('Day Type', **FONT_AXIS)
ax.set_ylabel('Cuisine Type', **FONT_AXIS)
plt.tight_layout()
plt.show()
```

**What this chart shows:** For each cuisine, what percentage of its orders come on weekdays vs. weekends.

**Why it matters:** Some cuisines may be heavily weekend-focused (leisure dining) while others spread more evenly (work-lunch culture). This shapes how restaurants should staff and how FoodHub should allocate delivery resources by cuisine zone.

**What FoodHub should do:** Use this to plan targeted weekend promotions for the most weekend-concentrated cuisines.

```python
# ── 8.3 Delivery Time Bottleneck Detection ────────────────────────────────────
# Define "bottleneck" orders: prep OR delivery time in the top 10th percentile
prep_p90 = df['food_preparation_time'].quantile(0.90)
del_p90  = df['delivery_time'].quantile(0.90)

slow_prep     = df[df['food_preparation_time'] > prep_p90]
slow_delivery = df[df['delivery_time'] > del_p90]
both_slow     = df[(df['food_preparation_time'] > prep_p90) & (df['delivery_time'] > del_p90)]

print('Operational Bottleneck Summary\n' + '─' * 40)
print(f'Preparation time 90th percentile : {prep_p90:.0f} min')
print(f'Delivery time 90th percentile    : {del_p90:.0f} min')
print(f'Orders with slow prep (>p90)     : {len(slow_prep):,}')
print(f'Orders with slow delivery (>p90) : {len(slow_delivery):,}')
print(f'Orders slow in BOTH phases       : {len(both_slow):,}')
```

```python
# Which cuisines appear most in bottleneck orders?
slow_cuisine_prep = slow_prep['cuisine_type'].value_counts().head(5)
slow_cuisine_del  = slow_delivery['cuisine_type'].value_counts().head(5)

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

slow_cuisine_prep.plot(kind='barh', ax=axes[0], color=ACCENT_COLOR)
axes[0].set_title('Top Cuisines in Slow Prep Orders', **FONT_TITLE)
axes[0].set_xlabel('Number of Orders')

slow_cuisine_del.plot(kind='barh', ax=axes[1], color=HIGHLIGHT)
axes[1].set_title('Top Cuisines in Slow Delivery Orders', **FONT_TITLE)
axes[1].set_xlabel('Number of Orders')

plt.tight_layout()
plt.show()
```

**What these charts show:** Which cuisines most frequently appear in orders that took an unusually long time to prepare or to deliver.

**Why it matters:** This directs the company to the specific cuisine categories (and by extension, specific restaurants) where operational intervention would have the greatest impact.

**What FoodHub should do:** Reach out to the restaurants associated with high-frequency slow preparation in these cuisines and work with them on kitchen efficiency or order prioritisation protocols.

## 9. Business Insights & Recommendations

### Q13: Conclusions and Strategic Recommendations

---

### Key Conclusions from the Analysis

**1. Demand is highly concentrated.**  
A small number of restaurants and cuisines drive the overwhelming majority of orders. Shake Shack, The Meatball Shop, and Blue Ribbon Sushi collectively represent a disproportionate share of platform activity. American, Japanese, and Italian cuisines similarly dominate. This creates dependency risk — any disruption to top partners has outsized impact on the business.

**2. Weekend demand creates operational pressure.**  
Orders are significantly skewed toward weekends. While the platform handles this reasonably well (delivery times only marginally longer on weekends), the gap will widen as order volume grows. Weekend operational capacity is a near-term constraint.

**3. The rating system is severely underutilised.**  
Nearly 38% of all orders receive no rating. Among the orders that are rated, the vast majority receive 4 or 5 stars, suggesting that silent customers are likely also satisfied — but the business cannot act on data it doesn't have. This is the single most actionable data quality problem in the dataset.

**4. Total fulfillment time is predictable — but 10% of orders breach the 60-minute threshold.**  
The mean total fulfillment time (prep + delivery) is approximately 51 minutes. Preparation time (~27 min) and delivery time (~24 min) are both consistent. However, 1 in 10 orders exceeds 60 minutes. These outlier orders are likely driving most of the poor ratings on the platform.

**5. The 25% commission tier is the revenue engine.**  
With ~59% of orders costing more than $20, the top commission tier generates the large majority of FoodHub's revenue. The business model is heavily dependent on higher-ticket orders. Strategies that grow average order value have outsized financial impact.

**6. Customer retention data suggests a long-tail distribution.**  
Many customers appear to order infrequently. A robust loyalty programme could improve lifetime value per customer significantly.

---

### Strategic Recommendations

#### 🏆 Restaurant Partnership & Quality
- **Tier your restaurant partnerships.** Top-10 revenue restaurants should receive dedicated account management, priority onboarding of new menu items, and quality assurance protocols.
- **Act on low-rated high-volume restaurants.** Any restaurant in the top-20 by order volume with an average rating below 4.0 should be flagged for an immediate quality review.
- **Expand the promotional offer programme.** Use the eligibility criteria (>50 ratings, avg > 4.0) as a public "FoodHub Certified Excellence" badge in the app to drive customer confidence.

#### ⏱️ Operational Efficiency
- **Introduce delivery time SLAs.** Set cuisine-specific and restaurant-specific preparation time benchmarks. Track violations weekly and share reports with partner restaurants.
- **Build a proactive communication system.** When an order is projected to exceed 60 minutes, automatically send the customer a push notification with a revised ETA. This dramatically reduces dissatisfaction from long waits.
- **Surge staffing on weekends.** Weekend order volume significantly exceeds weekday volume. Ensuring adequate delivery agent availability — particularly in the highest-density cuisine zones — will protect delivery times as the platform scales.

#### 💬 Customer Engagement & Ratings
- **Fix the rating gap urgently.** A 38% non-rating rate means the business is making restaurant quality decisions with incomplete data. Introduce a lightweight in-app rating prompt (one tap after delivery confirmation) and test a small incentive (e.g., 10 loyalty points) to increase response rate.
- **Segment satisfied-but-silent customers.** Use order history data to identify repeat customers who never rate. These are high-value customers whose feedback would be most informative.

#### 💰 Revenue Growth
- **Grow average order value.** Since ~59% of orders already exceed $20, strategies to push the remaining 41% above the threshold (upsell prompts, bundle offers, free delivery above $20) would significantly increase commission revenue with no acquisition cost.
- **Invest in American, Japanese, and Italian cuisine restaurant supply.** These three cuisines generate the most revenue. Adding more high-quality restaurant partners in these categories would directly grow the top line.
- **Build a loyalty programme.** High-frequency customers disproportionately drive revenue. A points-based system with redemption thresholds would increase order frequency, improve retention, and provide richer behavioural data.

#### 📊 Data & Analytics
- **Add timestamp granularity.** The dataset currently captures only aggregate timing. Recording the exact time-of-day for each order would unlock hour-by-hour demand analysis, enabling precision staffing and targeted promotions.
- **Track customer tenure.** Understanding how long a customer has been on the platform helps identify at-risk churners and enables personalised re-engagement.

## 10. Conclusion

This analysis of FoodHub's order data reveals a platform with strong operational foundations but clear, addressable gaps in customer engagement, data collection, and operational efficiency.

The business is healthy: delivery times are consistent, preparation times are predictable, and average order values support a viable commission structure. However, the platform is also highly concentrated — in restaurants, in cuisines, in days of the week, and in revenue tiers. This concentration creates both opportunity (focus resources on what matters most) and risk (vulnerability to disruption in a small number of key relationships).

The single most impactful near-term change would be to address the rating collection gap. Nearly 4 in 10 orders generate no customer feedback. Solving this problem would immediately improve the company's ability to monitor quality, reward excellent partners, and identify underperforming ones — all of which feed directly into customer satisfaction and long-term revenue growth.

The recommendations in this report are designed to be actionable, prioritised, and grounded directly in the data. They represent a roadmap for FoodHub to move from reactive operations to proactive, data-driven business management.
